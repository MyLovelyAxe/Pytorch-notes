## Building of Tensors

```python
import torch
import numpy as np
```

#### general method: torch.tensor( )

```python
'''torch.tensor()'''
# only permits 1 argument, which can be one single real number or one single list
# and directly transfer the input data in tensor type, without change anything else

# argument: one single real number, i.e. without []
# real num: integer
tensor0 = torch.tensor(7)
tensor1 = torch.tensor(0)
# real num: float
tensor2 = torch.tensor(7.)
tensor3 = torch.tensor(7.3)
tensor4 = torch.tensor(0.)

# argument: one single list, i.e. with []
# the number of outer [] equals dimension
# ps: whether argument elements are integer or float is similar with 'real num' argument
# however, if just one decimal element appears, all other elements will be float
# here to reduce the redundancy, only show 'integer' situation
# list: []
tensor5 = torch.tensor([7])
tensor6 = torch.tensor([7,8,9])
#list: [[]]
tensor7 = torch.tensor([[2,3]])
tensor8 = torch.tensor([[2,3,4],[4,5,6]])
#list: [[[]]]
tensor9 = torch.tensor([[[7]]])
tensor10 = torch.tensor([[[7,9,3],[4,3,0]],[[1,2,8],[5,6,4]]])
#list: [[[...]]] and so on in the similar fashion以此类推

lst_tensor = [tensor0, tensor1, tensor2, tensor3, tensor4, tensor5,
              tensor6, tensor7, tensor8, tensor9, tensor10]
for i,data in enumerate(lst_tensor):
    print(f'tensor {i}', ':', data)
    print('type:', data.type(),'\t','shape:', data.shape, '\t', 'dim:', data.dim())
    print('-------------------------------------------------------------')
```

#### directly with type: LongTensor, IntTensor, FloatTensor, DoubleTensor

```python
# directly create tensor with data type
# attention:
#    this kind of building is uninitialized
#    i.e. the initialization is purely random
#    values could be anyone from negative infinity to positive infinity

'''torch.LongTensor()'''
# due to the data type is specific, make sure use only integer as arguments
print('=======================================')
print('torch.LongTensor()')
print('=======================================')

# with [], then arguments are values of tensor
# num of [] equals dimension, similar to torch.tensor()
LT0 = torch.LongTensor([2])
print('LongTensor 0:', LT0)
print('type:', LT0.type(),'\t','shape:', LT0.shape, '\t', 'dim:', LT0.dim())
print('-------------------------------------------------------------')
LT1 = torch.LongTensor([[2,4,5],[3,3,3]])
print('LongTensor 1:', LT1)
print('type:', LT1.type(),'\t','shape:', LT1.shape, '\t', 'dim:', LT1.dim())
print('-------------------------------------------------------------')

# without [], then input is shape of tensor, random value
LT2 = torch.LongTensor(4)
LT3 = torch.LongTensor(2,3)
LT4 = torch.LongTensor(3,4,5)
LT5 = torch.LongTensor(6,3,8,8)

lst_LT = [LT2, LT3, LT4, LT5]
for i,data in enumerate(lst_LT):
    print(f'LongTensor {i+2}:')
    print('type:', data.type(),'\t','shape:', data.shape, '\t', 'dim:', data.dim())
    print('-------------------------------------------------------------')
print()
    
'''torch.IntTensor()'''
# basically same with torch.LongTensor()
print('=======================================')
print('torch.IntTensor()')
print('=======================================')

# with [] -> values
IT0 = torch.IntTensor([5,6])
print('IntTensor 0:', IT0)
print('type:', IT0.type(),'\t','shape:', IT0.shape, '\t', 'dim:', IT0.dim())
print('-------------------------------------------------------------')

# without [] -> shape, random values
IT1 = torch.IntTensor(3,4,2,2)
print('IntTensor 1:')
print('type:', IT1.type(),'\t','shape:', IT1.shape, '\t', 'dim:', IT1.dim())
print('-------------------------------------------------------------')
print()
    
'''torch.FloatTensor()'''
# arguments should be float
# others basically same with torch.LongTensor()
print('=======================================')
print('torch.FloatTensor()')
print('=======================================')

# with [] -> values
FT0 = torch.FloatTensor([2.,4.])
print('FloatTensor 0:', FT0)
print('type:', FT0.type(),'\t','shape:', FT0.shape, '\t', 'dim:', FT0.dim())
print('-------------------------------------------------------------')

# without [] -> shape, random values
FT1 = torch.FloatTensor(3,4,2,2)
print('FloatTensor 1:')
print('type:', FT1.type(),'\t','shape:', FT1.shape, '\t', 'dim:', FT1.dim())
print('-------------------------------------------------------------')
print()

'''torch.DoubleTensor()'''
# arguments should be float
# others basically same with torch.FloatTensor()
print('=======================================')
print('torch.DoubleTensor()')
print('=======================================')

# with [] -> values
DT0 = torch.DoubleTensor([7.8,9.])
print('DoubleTensor 0:', DT0)
print('type:', DT0.type(),'\t','shape:', DT0.shape, '\t', 'dim:', DT0.dim())
print('-------------------------------------------------------------')

# without [] -> shape, random values
DT1 = torch.DoubleTensor(4,3,6,6)
print('DoubleTensor 1:')
print('type:', DT1.type(),'\t','shape:', DT1.shape, '\t', 'dim:', DT1.dim())
print('-------------------------------------------------------------')
```

#### random initialization within certain scale

```python
'''torch.rand()'''
# the initialization of torch.rand() will be random float in: [0,1]
print('=======================================')
print('torch.rand()')
print('=======================================')
# arguments of torch.rand() can be:
#   tuple containing shape
rand_0 = torch.rand((3,3,3))
print('torch.rand((3,3,3)):')
print('type:', rand_0.type(),'\t','shape:', rand_0.shape, '\t', 'dim:', rand_0.dim())
print('----------------------------------------------------------------------------')
#   shape
rand_1 = torch.rand(3,3)
print('torch.rand(3,3):')
print('type:', rand_1.type(),'\t','shape:', rand_1.shape, '\t', 'dim:', rand_1.dim())
print('----------------------------------------------------------------------------')
#   shape with one-layer []
rand_2 = torch.rand([3])
print('torch.rand([3]):')
print('type:', rand_2.type(),'\t','shape:', rand_2.shape, '\t', 'dim:', rand_2.dim())
print('----------------------------------------------------------------------------')
print()
#   shape with one-layer []
rand_3 = torch.rand([3,3,3])
print('torch.rand([3,3,3]):')
print('type:', rand_3.type(),'\t','shape:', rand_3.shape, '\t', 'dim:', rand_3.dim())
print('----------------------------------------------------------------------------')
print()

'''torch.randn()'''
# initialization: random values subjective to normal distribution N~(0,1)
print('=======================================')
print('torch.randn()')
print('=======================================')
# argument choice is the same with torch.rand(): 
#    tupe of shape, shape, one-layer list of shape
randn_0 = torch.randn(3,3)
print('torch.tandn([3,3]):', randn_0)
print('type:', randn_0.type(),'\t','shape:', randn_0.shape, '\t', 'dim:', randn_0.dim())
print('----------------------------------------------------------------------------')
print()

'''torch.randint()'''
# arguments comprise: (min,max,(tuple_of_shape))
# return a tensor whose element values are in [min,max) with the given shape
# the shape should be given as a tuple or list
print('=======================================')
print('torch.randint()')
print('=======================================')
randint_0 = torch.randint(10,20,(3,3))
# same with randint_0 = torch.randint(10,20,[3,3])
print('torch.randint(1,10,(3,3)):', randint_0)
print('type:', randint_0.type(),'\t','shape:', randint_0.shape, '\t', 'dim:', randint_0.dim())
print('----------------------------------------------------------------------------')
print()
```

#### through other tensor: torch.X_like()

```python
# create an int module and a float module
m_int = torch.randint(1,10,(3,3))
print('m_int:', m_int)
print('type:', m_int.type(),'\t','shape:', m_int.shape, '\t', 'dim:', m_int.dim())
print('----------------------------------------------------------------------------')
m_float = torch.rand(3,3)
print('m_float:', m_float)
print('type:', m_float.type(),'\t','shape:', m_float.shape, '\t', 'dim:', m_float.dim())
print('----------------------------------------------------------------------------')
print()

'''torch.ones_like()'''
# get a same-shape tensor as argument whose elements are all 1
print('=======================================')
print('torch.ones_like()')
print('=======================================')
# without assign 'dtype'
# output will get a tensor with same shape and data type of argument
onelike_0 = torch.ones_like(m_int)
print('torch.ones_like(m_int):', onelike_0)
print('type:', onelike_0.type(),'\t','shape:', onelike_0.shape, '\t', 'dim:', onelike_0.dim())
print('----------------------------------------------------------------------------')
# without assign 'dtype'
# output will get a tensor with same shape and data type of argument
onelike_1 = torch.ones_like(m_float)
print('torch.ones_like(m_float):', onelike_1)
print('type:', onelike_1.type(),'\t','shape:', onelike_1.shape, '\t', 'dim:', onelike_1.dim())
print('----------------------------------------------------------------------------')
# assign 'dtype'
# dtype = torch.int / torch.long / torch.float / torch.double
# output will get a tensor with same shape of argument and data type of 'dtype'
onelike_2 = torch.ones_like(m_int, dtype = torch.double)
print('torch.ones_like(m_int, dtype = torch.double):', onelike_2)
print('type:', onelike_2.type(),'\t','shape:', onelike_2.shape, '\t', 'dim:', onelike_2.dim())
print('----------------------------------------------------------------------------')
print()

'''torch.zeros_like()'''
# get a same-shape tensor as argument whose elements are all 0
# others are totally same with torch.ones_like()

'''torch.rand_like()'''
# get a same-shape tensor as argument whose elements are in scale: [0,1]
print('=======================================')
print('torch.rand_like()')
print('=======================================')
# if argument is originally torch.DoubleTensor or torch.FloatTensor type
# everything is in order
randlike_0 = torch.rand_like(m_float)
print('torch.rand_like(m_float):', randlike_0)
print('type:', randlike_0.type(),'\t','shape:', randlike_0.shape, '\t', 'dim:', randlike_0.dim())
print('----------------------------------------------------------------------------')
# if argument is originally torch.LongTensor or torch.IntTensor type, report an error
# set dtype = torch.float or torch.double can solve this
randlike_1 = torch.rand_like(m_int, dtype = torch.float)
print('torch.rand_like(m_int, dtype = torch.float):', randlike_1)
print('type:', randlike_1.type(),'\t','shape:', randlike_1.shape, '\t', 'dim:', randlike_1.dim())
print('----------------------------------------------------------------------------')

'''torch.randn_like()'''
# get a same-shape tensor as argument whose elements are subjective to normal distribution N~(0,1)
# others are totally same with torch.rand_like()
```

#### torch.full( )

```python
'''torch.full()'''
# torch.full() return a tensor with desired shape full of the same desired values
# arguments: torch.full([shape],value)
#    [shape]: the shape of returned tensor, which should be a list
#    value:   the value that user wants to fulfill the tensor
full_0 = torch.full([2,3],7)
print('torch.full([2,3],7):', full_0)
print('type:', full_0.type(),'\t','shape:', full_0.shape, '\t', 'dim:', full_0.dim())
print('----------------------------------------------------------------------------')
# the type of tensor depends on the value desired
full_1 = torch.full([3],7.)
print('torch.full([3],7):', full_1)
print('type:', full_1.type(),'\t','shape:', full_1.shape, '\t', 'dim:', full_1.dim())
print('----------------------------------------------------------------------------')
# if the wanted is a scalar, just leave a blank []
# e.g. the following is the same with: torch.tensor(7)
full_2 = torch.full([],7)
# full_2 = torch.tensor(7)
print('torch.full([],7):', full_2)
print('type:', full_2.type(),'\t','shape:', full_2.shape, '\t', 'dim:', full_2.dim())
print('----------------------------------------------------------------------------')
```

#### similar with numpy

```python
'''torch.from_numpy()'''
#build data with numpy and transfer it into TorchTensor
np_data = np.ones([2,3])
print('np.ones([2,3]):', np_data)
np_like_0 = torch.from_numpy(np_data)
print('torch.from_numpy(np_data):', np_like_0)
print('type:', np_like_0.type(),'\t','shape:', np_like_0.shape, '\t', 'dim:', np_like_0.dim())
print('----------------------------------------------------------------------------')

'''torch.arange()'''
# arguments: torch.arange(min,max,step)
#    min:   the range starts from min(included)
#    max:   the range ends at max(excluded)
#    step:  the step of increment
np_like_1 = torch.arange(0,20,2)
print('torch.arange(0,20,2):', np_like_1)
print('type:', np_like_1.type(),'\t','shape:', np_like_1.shape, '\t', 'dim:', np_like_1.dim())
print('----------------------------------------------------------------------------')

'''torch.linspace()'''
# arguments: torch.linspace(min,max,num)
#    min:   the range starts from min(included)
#    max:   the range ends at max(included)
#    num:   the number of elements in the range with equally average intervals
np_like_2 = torch.linspace(0,20,11)
print('torch.linspace(0,20,10):', np_like_2)
print('type:', np_like_2.type(),'\t','shape:', np_like_2.shape, '\t', 'dim:', np_like_2.dim())
print('----------------------------------------------------------------------------')

'''torch.ones()'''
# arguments: torch.ones(shape) or torch.ones([shape_lst]) or torch.ones((shape_tuple))
np_like_3 = torch.ones([2,3])
print('torch.ones(2,3):', np_like_3)
print('type:', np_like_3.type(),'\t','shape:', np_like_3.shape, '\t', 'dim:', np_like_3.dim())
print('----------------------------------------------------------------------------')

'''torch.zeros()'''
# arguments: torch.zeros(shape) or torch.zeros([shape_lst]) or torch.zeros((shape_tuple))
np_like_4 = torch.zeros((2,3))
print('torch.zeros(2,3):', np_like_4)
print('type:', np_like_4.type(),'\t','shape:', np_like_4.shape, '\t', 'dim:', np_like_4.dim())
print('----------------------------------------------------------------------------')

'''torch.eye()'''
# return diagonal matrix
# arguments: torch.eye(shape)
# attention: torch.eye([shape]) doesn't work
# e.g. torch.eye(3,4), torch.eye(4), torch.eye(4,4) work
# however torch.eye([4,4]) doesn't work
np_like_5 = torch.eye(4,4)
print('torch.eye(2,3):', np_like_5)
print('type:', np_like_5.type(),'\t','shape:', np_like_5.shape, '\t', 'dim:', np_like_5.dim())
print('----------------------------------------------------------------------------')
```

#### torch.randperm()

```python
'''torch.randperm()'''
# torch.randperm() is used to generate random indices
# sometimes we need shuffle the data
# when there is more than 1 dataset to be shuffled and necessary to maintain correspondance
# we can create random indices, and shuffle different datesets with same indices

# arguments: torch.randperm(max)
#   max: returned random indices starts from 0 and ends at max(excluded)
rdp_0 = torch.randperm(10)
print('torch.randperm(10):', rdp_0)
print('type:', rdp_0.type(),'\t','shape:', rdp_0.shape, '\t', 'dim:', rdp_0.dim())
print('----------------------------------------------------------------------------')
```
