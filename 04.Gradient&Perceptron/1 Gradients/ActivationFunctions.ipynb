## Activation Functions and Gradients

```python
import torch
import matplotlib.pyplot as plt

def show_graph(data,act_func):
    if act_func == 'sigmoid':
        processed_data = torch.sigmoid(data)
    elif act_func == 'tanh':
        processed_data = torch.tanh(data)
    elif act_func == 'relu':
        processed_data = torch.relu(data)
    plt.figure(figsize=(6,4))
    plt.scatter(data,processed_data,marker='o',color='blue', label=act_func)
    plt.scatter(data,torch.zeros_like(data),marker='o',color='red', label="points")
    plt.xlabel('data')
    plt.ylabel('{} value'.format(act_func))
    plt.title('{} activation function'.format(act_func))
    plt.legend()
    plt.grid()
    plt.show()

data = torch.linspace(-10,10,41)
print(data)
```

### Sigmoid
<font size=2>

$$
\begin{equation}
\begin{aligned}
f(x) &= \sigma(x) = \frac{1}{1 + e^{-x}} \\
\frac{\partial{f(x)}}{\partial{x}} &= \sigma^{\prime} = \sigma(1 - \sigma)\\
\end{aligned}
\end{equation}
$$

```python
show_graph(data,'sigmoid')
```

### Tanh
<font size=2>

$$
\begin{equation}
\begin{aligned}
f(x) &= tanh(x) = \frac{e^{e}-e^{-x}}{e^{e}+e^{-x}} \\
&= 2\sigma(2x) - 1 \\
\frac{\partial{f(x)}}{\partial{x}} &= 1 - \left( \frac{e^{e}-e^{-x}}{e^{e}+e^{-x}} \right) = 1 - tanh^{2}(x) \\
\end{aligned}
\end{equation}
$$

```python
show_graph(data,'tanh')
```

### ReLU: Rectified Linear Unit
<font size=2>

$$
\begin{equation}
\begin{aligned}
f(x) = \begin{cases}
x,&x>0 \\ 0,&x<=0
\end{cases} \\
\frac{\partial{f(x)}}{\partial{x}} = \begin{cases}
1,&x>0 \\ 0,&x<=0
\end{cases} \\
\end{aligned}
\end{equation}
$$

```python
show_graph(data,'relu')
```
